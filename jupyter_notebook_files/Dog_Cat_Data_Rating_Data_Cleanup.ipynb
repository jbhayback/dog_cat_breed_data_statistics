```python
import pandas as pd

dog_cat_data_raw = pd.read_json('rating.json')
```

```python
cat_breed_data = dog_cat_data_raw[['cat_breeds']].dropna()
```

```python
cat = pd.DataFrame(list(cat_breed_data['cat_breeds']))
```

```python
cat['Breed'] = cat_breed_data.index
```

```python
cat.fillna(0, inplace=True)
```

```python
cat.columns = pd.Index([col.replace(' ', '_') for col in cat.columns])
```

```python
cat.to_csv('cat_breed_data.csv')
```

```python
dog_breed_data = dog_cat_data_raw[['dog_breeds']].dropna()
```

```python
dog = pd.DataFrame(list(dog_breed_data['dog_breeds']))
```

```python
dog['Breed'] = dog_breed_data.index
```

```python
dog.set_index('Breed', inplace=True)
```

```python
dog.dropna(inplace = True)
```

```python
dog.columns = pd.Index([col.replace(' ', '_') for col in dog.columns])
```

```python
dog.rename(columns={'_Adaptability': 'Adaptability', '_All_Around_Friendliness': 'All_Around_Friendliness', '_Exercise_Needs': 'Exercise_Needs', '_Health_Grooming':'Health_Grooming', '_Trainability':'Trainability'}, inplace=True)
```
